```python
#Import packages
import matplotlib.pyplot as plt
from samila import GenerativeImage
import random
import math
from samila import Projection
from samila import Marker
```

Image 1

```python
def f1(x, y): #Define function
    #Initialize and return result
    result = random.uniform(-1,7) * x**2  - math.sin(y**4) + abs(y-x)
    return result
def f2(x, y): #Define function
    #Initialize and return result
    result = random.uniform(-1,7) * y**2 - math.cos(x**4) + 2*x
    return result
g = GenerativeImage(f1, f2) #Initialize g as GenerativeImage
g.generate(seed=350) #Invoke generate method and set seed for replication
g.plot()
g.seed
plt.show()
```
