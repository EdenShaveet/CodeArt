```python
#Import packages
import matplotlib.pyplot as plt
from samila import GenerativeImage
import random
import math
from samila import Projection
from samila import Marker
```

```python
def f1(x, y): #Define function
    #Instantiate and return result
    result = random.uniform(-1,1) * x**1  - math.sin(y**1) + abs(y-x)
    return result
def f2(x, y): #Define function
    #Instantiate and return result
    result = random.uniform(-1,2) * y**2 - math.cos(x**2) + 2*x
    return result
g = GenerativeImage(f1, f2) #Instantiate g as GenerativeImage
g.generate(seed=20237) #Invoke generate method and set seed for replication
g.plot(projection=Projection.POLAR, marker=Marker.CIRCLE, spot_size=5, color="lightblue") #Plot image as blue polar plot with circle markers of size 10
g.seed
plt.show()
```
